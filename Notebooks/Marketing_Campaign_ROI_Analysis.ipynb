## 2. Key Performance Indicator (KPI) Overview

Our analysis involved calculating several critical KPIs to gain a holistic view of campaign efficacy:

*   **Click-Through Rate (CTR):** Measures the percentage of impressions that resulted in a click, indicating ad relevance and appeal.
*   **Conversion Rate:** Reflects the percentage of clicks that led to a desired action (conversion), highlighting the effectiveness of the conversion funnel.
*   **Return on Investment (ROI):** Quantifies the profitability of each campaign, measuring the financial gain relative to the cost. This is a primary metric for budget allocation decisions.
*   **Cost Per Conversion (CPCv):** Indicates the average cost incurred to achieve one conversion, crucial for evaluating campaign efficiency.

## 3. Performance Analysis

### 3.1. ROI Analysis: Identifying Value Drivers and Drainers

The distribution of ROI across campaigns reveals a varied landscape of performance. While some campaigns deliver exceptional returns, a significant portion shows negative or marginal ROI, indicating potential areas for re-evaluation.

These campaigns demonstrate exceptional efficiency in generating revenue relative to their cost. They represent prime candidates for increased investment.

**Underperforming Campaigns by ROI:**
